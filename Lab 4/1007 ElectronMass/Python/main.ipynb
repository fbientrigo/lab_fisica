```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
qm = pd.read_csv("qm_1.csv", dtype=np.float64)
print(qm)
```

```python
ax = qm.boxplot(figsize=(10,9))
plt.title(r'$\mu = \frac{q}{m}$ Electrón')

plt.ylabel(r'$q/m \times 10^{11} [C/kg]$')
plt.xlabel(r' Nº Dato')
plt.axhline(y = 1.75882017145, color = 'r', linestyle = '-', label=r"q/m literatura = $1.759$")
plt.axhline(y = 3.02711256373767, color = 'g', linestyle = '--', label=r"Mediana = $3.027$")


#plt.legend().remove() # se dejara legends para varios data
plt.legend()
plt.show()


```

# El error

$
\mu = \frac q m
$

$$
(\Delta \mu)^2 = (\frac{1}{B^3} \frac{2V}{R^2} \Delta B)^2 + (\frac{2}{(BR)^2} \Delta V)^2 + (\frac{1}{R^3} \frac{2V}{B^2} \Delta B)^2
$$



```python
qm = pd.read_csv("qm_1.csv", dtype=np.float64)
print(qm)

mediana = np.median(qm)

ax = qm.boxplot(figsize=(10,9))
plt.title(r'$\mu = \frac{q}{m}$ Electrón')

plt.ylabel(r'$q/m \times 10^{11} [C/kg]$')
plt.xlabel(r' Nº Dato')
plt.axhline(y = 1.75882017145, color = 'r', linestyle = '-', label=r"q/m literatura = $1.759$")
plt.axhline(y = mediana, color = 'g', linestyle = '--', label=f"Mediana = {mediana:.3}")


#plt.legend().remove() # se dejara legends para varios data
plt.legend()
plt.show()

```
